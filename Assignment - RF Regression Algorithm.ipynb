```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[["charges"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=50,random_state=0)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename='Finalized_Insurance_RF_Regresion.sav'
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("Finalized_Insurance_RF_Regresion.sav",'rb'))
```

```python
#input Parameter
age_input=float(input("Age :")) 
bmi_input=float(input("BMI :"))
child_input=float(input("Children:"))
sex_input=int(input("Sex 0 or 1 :"))
smok_input=int(input("Smoker 0 or 1 :"))
```

```python
result=loaded_model.predict([[age_input, bmi_input, child_input, sex_input, smok_input]])
print("Predicted Charges :",result)
```
